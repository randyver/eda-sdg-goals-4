```python
import pandas as pd
```

```python
df = pd.read_excel("4.5.xlsx")
```

```python
df[df.duplicated]
```

```python
df_indo = df[df['GeoAreaName'] == 'Indonesia']
```

```python
df_indo
```

```python
df_indo.columns
```

```python
df_indo.drop(['Goal', 'Target', 'Indicator', 'SeriesCode', 'SeriesDescription', 'GeoAreaCode', 'GeoAreaName', 'Reporting Type'], axis = 1,inplace= True)
```

```python
df_indo
```

```python
df_indo.drop('Units', axis=1, inplace= True)
```

```python
df_indo
```

```python
df_indo.drop(['2001','2007','2013','2016','2017','2019','2021'], axis= 1,inplace=True)
```

```python
df_indo
```

```python
import matplotlib.pyplot as plt
import seaborn as  sns
df_long = pd.melt(df_indo, id_vars=["Education level", "Type of skill"], 
                  var_name="Year", value_name="Value")


df_long["Year"] = pd.to_numeric(df_long["Year"])


plt.figure(figsize=(10, 6))
sns.lineplot(data=df_long, x="Year", y="Value", hue="Type of skill", style="Education level", markers=True)
plt.title("Skill Progression by Education Level (2000-2018)")
plt.ylabel("Value")
plt.xlabel("Year")
plt.xticks(df_long["Year"].unique())  
plt.show()
```

Dari visualisasi di atas artinya seiring majunya tahun rasio tingkat kemahiran minimum di bidang membaca dan matematika menurun yang artinya pendidikan di perkotaan maju(menjadi lebih baik) lebih signifikan dari pada pendidikan di desa. rasio pada tahun 2018 juga terlihat sangat rendah yaitu sekitar 0.3-0.4 yang artinya capaian tingkat kemahiran minimum di bidang membaca dan matematika di desa sangat jauh di bawah capaian tingkat kemahiran minimum di bidang membaca dan matematika di kota.
