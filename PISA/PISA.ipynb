```python
import pandas as pd

data = pd.read_excel("PISA-INDO.xlsx")
data.head()
```

```python
data.drop(columns='No', axis=1,inplace=True)
```

```python
data.head()
```

```python
data.replace({'Nama Data':"Tahun"})
```

```python
data.rename(columns={'Nama Data': 'Tahun'},inplace=True)
```

```python
data
df = data
```

```python
import matplotlib.pyplot as plt

df_filtered= df

plt.figure(figsize=(10,6))
plt.plot(df_filtered['Tahun'], df_filtered['Matematika'], marker='o', label='Matematika')
plt.plot(df_filtered['Tahun'], df_filtered['Membaca'], marker='o', label='Membaca')

# Add labels and title
plt.xlabel('Tahun')
plt.ylabel('Nilai')
plt.title('Perkembangan Nilai PISA Matematika dan Membaca Indonesia')
plt.legend()

# Add the actual values on top of the points
for i in range(len(df_filtered)):
    plt.text(df_filtered['Tahun'].iloc[i], df_filtered['Matematika'].iloc[i] + 1, f"{df_filtered['Matematika'].iloc[i]:.2f}", ha='center')
    plt.text(df_filtered['Tahun'].iloc[i], df_filtered['Membaca'].iloc[i] + 1, f"{df_filtered['Membaca'].iloc[i]}", ha='center')
   
# Show the plot
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()

```
