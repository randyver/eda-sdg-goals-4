```python
df_indo = df[df['GeoAreaName'] == "China, Hong Kong Special Administrative Region"]
df_indo
```

```python
# Pilih kolom dari tahun 2005 hingga 2023
df_indo_time = df_indo.loc[:, '2005':'2023']
```

```python
df_indo['SeriesDescription'].value_counts()
```

```python
df_indo_filtered = df_indo[['Education level', 'Sex', '2022']]
df_indo_filtered
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan dataframe Anda bernama df
sns.barplot(x="Education level", y="2022", hue="Sex", data=df_indo_filtered)
plt.title("2022 Values per Education Level and Sex")
plt.show()
```
