```python
# read data from xlsx
import pandas as pd

df = pd.read_csv('sma_total.csv')
df.head(50)
```

```python
import matplotlib.pyplot as plt

# Hitung total setiap tahun
total_per_tahun = df[['2021', '2022', '2023']].sum()

# Buat grafik line chart
plt.plot(total_per_tahun.index, total_per_tahun.values, marker='o')

# Tambahkan value eksak di setiap tahun dengan warna oranye
# for i, value in enumerate(total_per_tahun.values):
#     plt.text(total_per_tahun.index[i], value, f'{value:.0f}', ha='center', va='bottom', color='green')

# Judul dan label
plt.title('Total Angka Putus Sekolah Tingkat SMA')
plt.xlabel('Tahun')
plt.ylabel('Total')
plt.grid(True)

# Tampilkan grafik
plt.show()
```

```python
# Membuat stacked bar chart untuk perbandingan jumlah setiap tahun per provinsi
df.set_index('Provinsi')[['2021', '2022', '2023']].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Pertumbuhan Jumlah per Provinsi Angka Putus Sekolah Tingkat SMA (2021-2023)')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table


# Menyiapkan figure
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# Daftar tahun yang akan dianalisis
years = ['2021', '2022', '2023']

# Membuat tabel untuk setiap tahun
for i, year in enumerate(years):
    # Mengurutkan data berdasarkan tahun dan memilih 5 provinsi dengan angka putus sekolah terbanyak
    top_provinces = df.nlargest(5, year)
    
    # Menampilkan tabel di subplot
    axes[i].axis('off')  # Menyembunyikan axis
    table_data = top_provinces[['Provinsi', year]].rename(columns={year: 'Angka Putus Sekolah'})
    table_plot = table(axes[i], table_data, loc='center', cellLoc='center', colWidths=[0.2, 0.2])
    
    # Menambahkan judul
    axes[i].set_title(f'5 Provinsi Terbesar {year}')

# Menambahkan judul umum
plt.suptitle('5 Provinsi Terbesar Berdasarkan Angka Putus Sekolah Tingkat SMA', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Menyesuaikan layout agar tidak overlap dengan judul

# Menampilkan plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table


# Menyiapkan figure
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# Daftar tahun yang akan dianalisis
years = ['2021', '2022', '2023']

# Membuat tabel untuk setiap tahun
for i, year in enumerate(years):
    # Mengurutkan data berdasarkan tahun dan memilih 5 provinsi dengan angka putus sekolah terkecil
    smallest_provinces = df.nsmallest(5, year)
    
    # Menampilkan tabel di subplot
    axes[i].axis('off')  # Menyembunyikan axis
    table_data = smallest_provinces[['Provinsi', year]].rename(columns={year: 'Angka Putus Sekolah'})
    table_plot = table(axes[i], table_data, loc='center', cellLoc='center', colWidths=[0.2, 0.2])
    
    # Menambahkan judul
    axes[i].set_title(f'5 Provinsi Terkecil {year}')

# Menambahkan judul umum
plt.suptitle('5 Provinsi Terkecil Berdasarkan Angka Putus Sekolah Tingkat SMA', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Menyesuaikan layout agar tidak overlap dengan judul

# Menampilkan plot
plt.show()
```
