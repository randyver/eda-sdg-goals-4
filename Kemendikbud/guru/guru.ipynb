```python
import matplotlib.pyplot as plt

# Data
years = ['2015', '2020', '2021']
values = [56.69, 53.02, 50.64]

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(years, values, color='blue')

# Add line chart on top of the bar chart
plt.plot(years, values, color='red', marker='o', linewidth=2, markersize=8)

# Add title and labels
plt.title('Rerata Nilai Uji Kompetensi Guru (UKG)', fontsize=18)
plt.xlabel('Tahun', fontsize=16)
plt.ylabel('Nilai', fontsize=16)

# Set x-axis labels (years) font size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Show the plot
plt.show()
```
