```python
# read data
import pandas as pd

df = pd.read_excel('4.a.xlsx')
df.head()
```

```python
df.nunique()
```

```python
# ambil column selain column yang memiliki unique value 1 atau 0
df = df.loc[:, df.nunique() > 1]
df.nunique()
```

```python
# pilih kolom yang diperlukan saja
df = df[["SeriesCode", "SeriesDescription", "GeoAreaCode", "GeoAreaName", "TimePeriod", "Value", "Education level"]]
df.head()
```

```python
df_indonesia = df[df["GeoAreaName"] == "Indonesia"]
df_indonesia.info()
```

```python
df_indonesia.head(20)
```

```python
# isi dari kolom SeriesDescription
df_indonesia["SeriesDescription"].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get unique SeriesDescription and Education level combinations
unique_series = df_indonesia['SeriesDescription'].unique()
unique_education_levels = df_indonesia['Education level'].unique()

# Set the style for seaborn
sns.set(style="whitegrid")

# Loop through each unique SeriesDescription and Education level combination
for series_description in unique_series:
    for education_level in unique_education_levels:
        # Filter the data for this specific combination
        filtered_data = df_indonesia[
            (df_indonesia['SeriesDescription'] == series_description) &
            (df_indonesia['Education level'] == education_level)
        ]
        
        if not filtered_data.empty:
            # Create the plot
            plt.figure(figsize=(10, 6))
            sns.lineplot(
                data=filtered_data,
                x='TimePeriod', y='Value', marker='o'
            )

            # Add title and labels
            plt.title(f'{series_description} ({education_level})')
            plt.xlabel('Year')
            plt.ylabel('Value (%)')

            # Show the plot
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
```
